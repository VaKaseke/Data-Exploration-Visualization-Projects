```python
#Importing Required libraries 
# Libraries for data manipulation
import pandas as pd
import numpy as np

# Libraries for data visualisation
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

```python
# apply some unique styling
plt.style.use("ggplot")
rcParams['figure.figsize'] = (12,  6)
```

```python
# use sklearn to import a dataset
from sklearn.datasets import load_wine
```

```python
# Importing the dataset
wine = load_wine()

# Converting the dataset into a Pandas dataframe
df = pd.DataFrame(data=wine.data, columns=wine.feature_names)
```

```python
# Creating a column for the target
df["target"] = wine.target
```

```python
#View top rows to get big picture
df.head()
```

```python
#View bottom 5 rows
df.tail()
```

```python
#View the dimensionality of our dataset
df.shape
```

```python
#Descriptive information about dataset
df.describe()
```

```python
#Shorter summary of dataset including data types, null values and memory usage
df.info()
```

```python
#Data Preparation
#To check for duplicate rows
df.duplicated().sum()
```

```python
#Renaming unclear column od280 / od315_of_diluted_wines
df.rename(columns={"od280/od315_of_diluted_wines": "protein_concentration"}, inplace=True)
```

```python
#Understanding the variables
#categorical variables
df.target.value_counts()
```

```python
#Express the data as a percentage
df.target.value_counts(normalize=True)
```

```python
#Plot the data
df.target.value_counts().plot(kind="bar")
plt.title("Value counts of the target variable")
plt.xlabel("Wine type")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
#Analyse numeric variables
df.magnesium.describe()
```

```python
#Plot distribution of numeric variable
df.magnesium.plot(kind="hist")
plt.title("Distribution of Magnesium")
plt.xlabel("Values")
plt.xticks(rotation=0)
plt.ylabel("Frequency")
plt.show()
```

```python
#evaluate distribution kurtosis and asymmetry

print(f"Skewness: {df['magnesium'].skew()}")
print(f"Kurtosis: {df['magnesium'].kurt()}")
```

```python
#The distribution:

#does not follow a normal curve
#show spikes
#has kurtosis and asymmetry values greater than 1
```

```python
#Exploring relationships between variables
#Seaborn Pairplot useful for grasping the most important relationships without having to go through every single combination manually
sns.pairplot(df)
```

```python
#Boxplot to understand the relationship between numeric and categorical variables
#alcohol, flavanoids, color_intensity and proline visually show slightly more marked segmentations for a given wine type
sns.catplot(x="target", y="proline", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs proline")
plt.show()
```

```python
#The proline median of type 0 wine is bigger than that of the other two types
```

```python
#For flavanoids
sns.catplot(x="target", y="flavanoids", data=df, kind="box", aspect=1.5)
plt.title("Boxplot for target vs flavanoids")
plt.show()
```

```python
#Here too, the type 0 wine seems to have higher values of flavanoids
```

```python
#To investigate whether type 0 wines have higher combined levels of proline and flavanoids, create a scatterplot and visualize which wine class a point belongs to, specifying the hue parameter.

sns.scatterplot(x="proline", y="flavanoids", hue="target", data=df, palette="Dark2", s=80)
plt.title("Relationship between proline, flavanoids and target")
plt.show()
```

```python
#Conclude that Type 0 wines show clear patterns of flavanoids and proline. 
#In particular, the proline levels are much higher while the flavanoid level is stable around the value of 3
```

```python
#Use heatmap to expand exploration
# heatmap allows us to efficiently grasp which variables are strongly correlated with each other
corrmat = df.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=df.columns, 
                 xticklabels=df.columns, 
                 cmap="Spectral_r")
plt.show()
```

```python
#From above, conclude that when the target variable decreases (tends towards 0 or Type of wine 0) the flavanoids, total phenols, proline and other proteins tend to increase. 
#Also observe there is a very strong correlation between alcohol and proline
```
